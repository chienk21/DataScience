```python
import numpy as np
import csv
import numpy

def readCSVToNumpyArray(dataset):
    values = [[]]
    with open(dataset) as f:
        counter = 0
        for i in csv.reader(f):
            for j in i:
                try:
                    values[counter].append(float(j))
                except ValueError:
                    values[counter].append(j)
            counter = counter + 1
            values.append([])

    data = numpy.array(values[:-1],dtype='object')
    firstFiveData = numpy.array(values[1:6],dtype='object')
    
    for i in range(len(data[:-1])):
            CountryRegion = data[:,3]
            confirmedCases = data[:,7]
            deaths = data[:,8]
            recovered = data[:,9]
            activeCases = data[:,10] 
            #print(CountryRegion[i+1])
            #print(previousRegion)
            if CountryRegion[i+1] == CountryRegion[i]:
                RealCountryRegion = []                
                for i in CountryRegion:
                    if i not in RealCountryRegion:
                        RealCountryRegion.append(i)

                realConfirmedCases = confirmedCases[i+1]  + confirmedCases[i+2]
                #print(realConfirmedCases)
                print(str(RealCountryRegion) + '\'s' + ' confirmed cases: ' + str(realConfirmedCases))
            else:
                print(str(CountryRegion[i+1]) + '\'s' + ' confirmed cases: ' + str(confirmedCases[i+1]) + ', deaths: ' + str(deaths[i+1]) + ', recovered: ' + str(recovered[i+1]) + ', active cases: ' + str(activeCases[i+1]))
            
    return data

numpyArray = readCSVToNumpyArray('01-01-2021.csv')
```
